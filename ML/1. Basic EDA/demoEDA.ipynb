# Univariate Analysis

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

## 1. Categorical Data - Count Plot

```python
sns.countplot(x = df["Sex"])
```

```python
sns.countplot(x = df['Pclass'])
```

## 2. Categorical Data - Pie Chart

```python
df['Sex'].value_counts()
```

```python
df['Sex'].value_counts().plot(kind = 'pie')
```

```python
df["Sex"].value_counts().plot(kind="pie", autopct = '%.2f')
```

## Numerical Data - Histogram

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['Age'],bins=10)
```

```python
sns.histplot(df["Age"],kde=True)
```

```python
sns.boxplot(df['Age'])
```

# Bivariate and Multivariate Analysis

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

```python
tips.head()
```

```python
titanic = pd.read_csv('train.csv')
```

```python
flights.head()
```

## Numerical - Numerical

```python
sns.scatterplot(x = tips['total_bill'],y = tips['tip'])
```

```python
sns.scatterplot(x=tips["total_bill"], y=tips["tip"],hue=tips['sex'],style=tips['smoker'])
```

```python
sns.barplot(x = titanic['Pclass'],y = titanic['Age'])
```

```python
sns.barplot(x=titanic["Pclass"], y=titanic["Age"],hue=titanic['Sex'])
```

## Numerical - Categorical

```python
sns.boxplot(x = titanic['Sex'],y = titanic['Age'])
```

```python
cross = pd.crosstab(titanic["Pclass"], titanic["Survived"])
print(cross)
```

```python
sns.heatmap(cross)
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue="species")
```
